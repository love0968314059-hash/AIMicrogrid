## 3. 启动系统

```python
from colab_main import run_in_colab

# 在后台线程中运行
import threading
server_thread = threading.Thread(target=run_in_colab, daemon=True)
server_thread.start()

# 等待服务器启动
import time
time.sleep(5)

print("系统已启动！")
```

## 4. 使用ngrok暴露端口（如果需要）

```python
from pyngrok import ngrok

# 暴露5000端口
public_url = ngrok.connect(5000)
print(f"公共访问地址: {public_url}")
print(f"请在浏览器中打开: {public_url}")
```

## 5. 测试系统

```python
import requests

# 测试API
response = requests.get('http://localhost:5000/api/status')
print("系统状态:")
print(response.json())
```
